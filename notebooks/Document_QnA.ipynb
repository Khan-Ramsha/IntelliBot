```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.llms import huggingface_hub,HuggingFaceHub
from langchain.chains import LLMChain
```

```python
from langchain_community.document_loaders import PyPDFLoader
mypdf = PyPDFLoader('Artificial_Intelligence.pdf')
```

```python
content = mypdf.load()
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 200)
```

```python
chunk_doc=splitter.split_documents(content)
```

```python
chunk_doc
```

```python
embedding_model = HuggingFaceEmbeddings()
vector_store = FAISS.from_documents(documents = chunk_doc, embedding = embedding_model)
retriever = vector_store.as_retriever()
```

```python
vector_store
```

```python
retriever
```

```python
pipeline = HuggingFaceHub(
    repo_id="google/gemma-2-2b-it",
    model_kwargs={"temperature": 0.7},
    huggingfacehub_api_token="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"
)
```

```python
qa = RetrievalQA.from_chain_type(llm=pipeline, retriever=retriever, chain_type="stuff")

query = "What is AI?"

answer = qa.run(query)
```
